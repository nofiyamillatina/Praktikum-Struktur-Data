```python
# Implementasi Queue menggunakan List
class QueueList:
  def __init__(self):
    self.queue = []

  def enqueue(self, item):
    self.queue.append(item)

  def dequeue(self):
    if not self.is_empty():
      return self.queue.pop(0)
    return None

  def front(self):
    return self.queue[0] if not self.is_empty() else None

  def rear(self):
    return self.queue[-1] if not self.is_empty() else None

  def is_empty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)


# Contoh penggunaan
queue = QueueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
from collections import deque

# Implementasi Queue menggunakan deque
class QueueDeque:
  def __init__(self):
    self.queue = deque()

  def enqueue(self, item):
    self.queue.append(item)

  def dequeue(self):
    return self.queue.popleft() if not self.is_empty () else None

  def front(self):
    return self.queue[0] if not self.is_empty() else None

  def rear(self):
    return self.queue[-1] if not self.is_empty() else None

  def is_empty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)


# Contoh penggunaan
queue = QueueDeque()
queue.enqueue(40)
queue.enqueue(50)
queue.enqueue(60)
print("Queue setelah enqueue:", list(queue.queue))
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class QueueLinkedList:
  def __init__(self):
    self.front = self.rear = None

  def enqueue(self, data):
    new_node = Node(data)
    if self.rear is None:
      self.front = self.rear = new_node
      return
    self.rear.next = new_node
    self.rear = new_node

  def dequeue(self):
    if self.front is None:
      return None
    dequeued_element = self.front.data
    self.front = self.front.next
    if self.front is None:
      self.rear = None
    return dequeued_element

  def is_empty(self):
    return self.front is None

  def front_element(self):
    return self.front.data if self.front else None

  def rear_element(self):
    return self.rear.data if self.rear else None


# Contoh penggunaan
queue = QueueLinkedList()
queue.enqueue(40)
queue.enqueue(50)
queue.enqueue(60)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front_element())
print("Rear element:", queue.rear_element())
```

# Latihan 1

Deskripsi:

Buatlah sistem antrian pemesanan tiket bioskop menggunakan Queue berbasis List, di mana
setiap pelanggan yang datang akan mendapatkan nomor antrian dan dilayani berdasarkan
urutan kedatangan.

Ketentuan:

1. Gunakan dua huruf pertama nama mahasiswa + dua digit terakhir NPM sebagai kode
pelanggan.

2. Sistem harus memiliki metode untuk:

3. enqueue(ticket_id) → Menambahkan pelanggan ke dalam antrian.

4. dequeue() → Melayani pelanggan pertama dalam antrian.

5. front() dan rear() → Menampilkan pelanggan pertama dan terakhir dalam antrian.

6. is_empty() dan size() → Mengecek apakah antrian kosong dan menghitung jumlah
pelanggan dalam antrian.

Tambahan:
1. Jika antrian mencapai kapasitas maksimal (misalnya 10 pelanggan), pelanggan baru
harus menunggu hingga ada slot kosong.

2. Buat fitur "VIP Customer", di mana pelanggan dengan kode khusus (misalnya kode
pelanggan yang berisi angka 7) dapat langsung mendapatkan kursi tanpa harus
mengantri.

```python
class AntrianBioskop:
    def __init__(self, kapasitas=10):
        self.antrian = []
        self.kapasitas = kapasitas

    def enqueue(self, kode_pelanggan):
        if self.size() < self.kapasitas:
            self.antrian.append(kode_pelanggan)
            print(f"Pelanggan {kode_pelanggan} ditambahkan ke antrian.")
        else:
            print("Maaf, antrian penuh. Silakan tunggu hingga ada slot kosong.")

    def dequeue(self):
        if not self.is_empty():
            pelanggan = self.antrian.pop(0)
            print(f"Pelanggan {pelanggan} dilayani.")
            return pelanggan
        else:
            print("Antrian kosong.")
            return None

    def front(self):
        if not self.is_empty():
            return self.antrian[0]
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.antrian[-1]
        else:
            return None

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

    def tambah_pelanggan_vip(self, kode_pelanggan):
        print(f"Pelanggan VIP {kode_pelanggan} boleh masuk!")


# Contoh penggunaan
antrian = AntrianBioskop()

while True:
    print("\nMenu:")
    print("1. Tambah Pelanggan")
    print("2. Layani Pelanggan")
    print("3. Tambah Pelanggan VIP")
    print("4. Lihat Antrian")
    print("5. Keluar")

    pilihan = input("Masukkan pilihan: ")

    if pilihan == "1":
        nama = input("Masukkan nama pelanggan: ")
        npm = input("Masukkan NPM pelanggan: ")
        kode_pelanggan = nama[:2] + npm[-2:]
        antrian.enqueue(kode_pelanggan)
    elif pilihan == "2":
        antrian.dequeue()
    elif pilihan == "3":
        nama = input("Masukkan nama pelanggan VIP: ")
        npm = input("Masukkan NPM pelanggan VIP: ")
        kode_pelanggan = nama[:2] + npm[-2:]
        antrian.tambah_pelanggan_vip(kode_pelanggan)
    elif pilihan == "4":
        print(f"Pelanggan pertama: {antrian.front()}")
        print(f"Pelanggan terakhir: {antrian.rear()}")
        print(f"Ukuran antrian: {antrian.size()}")
        print(f"Antrian kosong? {antrian.is_empty()}")
    elif pilihan == "5":
         print("Terima Kasih! Sampai Jumpa Kembali di Bioskop Kami.")
         break
    else:
        print("Pilihan tidak valid.")
```

# Latihan 2

Deskripsi:

Buatlah sistem antrian layanan fotokopi di kampus menggunakan Queue berbasis List, di
mana mahasiswa yang datang lebih dulu akan lebih dulu mendapatkan layanan.

Ketentuan:

1. Gunakan tiga huruf pertama nama mahasiswa + dua digit terakhir NPM sebagai ID
pemesanan fotokopі.

2. Sistem harus memiliki metode untuk:

3. enqueue(kode_pesanan, Jml_halaman) → Menambahkan pesanan fotokopi ke dalam antrian
dengan jumlah halaman tertentu.

4. dequeue() → Melayani pesanan pertama dalam antrian.
front() dan rear() → Menampilkan pesanan pertama dan terakhir dalam antrian.

5. is_empty() dan size() → Mengecek apakah antrian kosong dan menghitung jumlah
pesanan dalam antrian.

Tambahan:

1. Jika pesanan memiliki lebih dari 100 halaman, pesanan tersebut akan diproses dalam
dua tahap, sehingga tetap berada di antrian setelah tahap pertama selesai.

2. Buat fitur "Fast Track", di mana pesanan kurang dari 10 halaman langsung diproses
tanpa harus mengantri.

```python
class AntrianFotokopi:
    def __init__(self):
        self.antrian = []

    def enqueue(self, kode_pesanan, Jml_halaman):
        self.antrian.append((kode_pesanan, Jml_halaman))
        print(f"Pesanan {kode_pesanan} dengan jumlah pesanan sebanyak {Jml_halaman} halaman ditambahkan ke antrian.")

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        kode_pesanan, Jml_halaman = self.antrian.pop(0)
        print(f"Memproses pesanan {kode_pesanan} dengan jumlah pesanan sebanyak {Jml_halaman} halaman.")
        if Jml_halaman > 100:
            self.enqueue(kode_pesanan, Jml_halaman - 100)
            print(f"Pesanan {kode_pesanan} dengan sisa {Jml_halaman - 100} halaman ditambahkan kembali ke antrian.")
        return kode_pesanan, Jml_halaman

    def front(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        return self.antrian[0]

    def rear(self):
        if self.is_empty():
            print("Antrian kosong.")
            return None
        return self.antrian[-1]

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

    def fast_track(self, kode_pesanan, Jml_halaman):
        if Jml_halaman < 10:
            print(f"Pesanan {kode_pesanan} dengan jumlah pesanan sebanyak {Jml_halaman} halaman diproses langsung (Fast Track).")
            return True
        return False

def generate_kode_pesanan(nama, npm):
    """Menghasilkan ID pesanan dari nama dan NPM."""
    nama_3_huruf = nama[:3].upper()
    npm_2_digit = npm[-2:]
    return nama_3_huruf + npm_2_digit

# Contoh penggunaan dengan input pengguna
antrian = AntrianFotokopi()

while True:
    print("\nMenu:")
    print("1. Tambah Pesanan (Enqueue)")
    print("2. Proses Pesanan (Dequeue)")
    print("3. Fast Track (Pesanan Cepat)")
    print("4. Lihat Pesanan Pertama")
    print("5. Lihat Pesanan Terakhir")
    print("6. Lihat Jumlah Pesanan")
    print("7. Keluar")

    pilihan = input("Masukkan pilihan Anda: ")

    if pilihan == "1":
        nama = input("Masukkan nama mahasiswa: ")
        npm = input("Masukkan NPM mahasiswa: ")
        kode_pesanan = generate_kode_pesanan(nama, npm)
        Jml_halaman = int(input("Masukkan jumlah halaman: "))
        antrian.enqueue(kode_pesanan, Jml_halaman)
    elif pilihan == "2":
        antrian.dequeue()
    elif pilihan == "3":
        nama = input("Masukkan nama mahasiswa: ")
        npm = input("Masukkan NPM mahasiswa: ")
        kode_pesanan = generate_kode_pesanan(nama, npm)
        Jml_halaman = int(input("Masukkan jumlah halaman: "))
        antrian.fast_track(kode_pesanan, Jml_halaman)
    elif pilihan == "4":
        print(f"Pesanan pertama: {antrian.front()}")
    elif pilihan == "5":
        print(f"Pesanan terakhir: {antrian.rear()}")
    elif pilihan == "6":
        print(f"Jumlah pesanan dalam antrian: {antrian.size()}")
    elif pilihan == "7":
        print("Terima kasih telah menggunakan layanan fotokopi kami.")
        break
    else:
        print("Pilihan tidak valid.")
```

# Latihan 3

Deskripsi:

Buatlah sistem antrian kendaraan masuk ke parkiran mall menggunakan Queue berbasis List, di mana kendaraan masuk dan keluar berdasarkan urutan kedatangannya.

Ketentuan:

1. Gunakan 2 huruf pertama kota lahir + 2 digit terakhir NPM sebagai nomor plat kendaraan.
2. Sistem harus memiliki metode untuk:
3. enqueue(vehicle_number) → Menambahkan kendaraan ke dalam antrian parkir.
4. dequeue() → Mengeluarkan kendaraan pertama dan terakhir dalam antrian.
5. is_empty() dan size() → Mengecek apakah parkiran penuh dan menghitung jumlah kendaraan dalam antrian.

Tambahan:

1. Jika jumlah kendaraan melebihi kapasistas parkiran (misalnya 15 kendaraan), kendaraan baru harus menunggu hingga ada kendaraan yang keluar.
2. Buat fitur "Parkir Prioritas", di aman kendaraan ramah lingkungan(misalnya kendaraan listrik) mendapatkan prioritas masuk parkiran lebih dulu.


```python
class AntrianParkir:
    def __init__(self, kapasitas=15):
        self.antrian = []
        self.kapasitas = kapasitas

    def enqueue(self, nomor_plat, jenis_kendaraan="biasa"):
        if self.size() < self.kapasitas:
            if jenis_kendaraan == "listrik":
                self.antrian.insert(0, nomor_plat)
                print(f"Kendaraan {nomor_plat} (listrik) masuk parkiran.")
            else:
                self.antrian.append(nomor_plat)
                print(f"Kendaraan {nomor_plat} masuk parkiran.")
        else:
            print(f"Parkiran penuh, kendaraan {nomor_plat} harus menunggu.")

    def dequeue(self):
        if not self.is_empty():
            kendaraan_keluar = self.antrian.pop(0)
            print(f"Kendaraan {kendaraan_keluar} keluar dari parkiran.")
        else:
            print("Parkiran kosong.")

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

    def tampilkan_antrian(self):
        if not self.is_empty():
            print("Antrian kendaraan di parkiran:", self.antrian)
        else:
            print("Parkiran kosong.")

# Fungsi untuk menghasilkan nomor plat
def buat_nomor_plat(kota, npm):
    return kota[:2] + npm[-2:]

# Contoh penggunaan dengan input pelanggan
parkiran = AntrianParkir()

while True:
    print("\nMenu:")
    print("1. Masuk Parkiran")
    print("2. Keluar Parkiran")
    print("3. Tampilkan Antrian")
    print("4. Keluar Program")

    pilihan = input("Masukkan pilihan Anda: ")

    if pilihan == "1":
        kota = input("Masukkan kota lahir Anda: ")
        npm = input("Masukkan NPM Anda: ")
        jenis_kendaraan = input("Jenis kendaraan (biasa/listrik): ").lower()
        nomor_plat = buat_nomor_plat(kota, npm)
        parkiran.enqueue(nomor_plat, jenis_kendaraan)
    elif pilihan == "2":
        parkiran.dequeue()
    elif pilihan == "3":
        parkiran.tampilkan_antrian()
    elif pilihan == "4":
        break
    else:
        print("Pilihan tidak valid.")

print("Selamat Datang Kembali!")
```

# Latihan 4


Deskripsi:

Buatlah sistem antrian pengambilan sertifikat di akmpus menggunakan Queue berbasis List, di mana mahasiswa yang telah menyelesaikan seminar atau pelatihan dapat mengambil sertifikat berdasarkan urutan kedatangan.

Ketentuan:

1. Gunakan huruf pertama program studi + 2 digit terakhir NPM sebagai nomor antrian mahasiswa.
2. Sistem harus memiliki metode untuk:
3. enqueue(student_id)  → Menambahkan mahasiswa ke dalam antrian pengambilan sertifikat.
4. dequeue() → Melayani mahasiswa yang lebih dulu datang.
front() dan rear()  → Menampilkan mahasiswa pertama dan terakhir dalam antrian.
5. is_empty() dan size()  → Mengecek apakah antrian kosong dan menghitung jumlah mahasiswa dalam antrian.

Tambahan:

1. Jika jumlah mahasiswa dalam antrian melebihi 20 orang, antrian akan dibagi menjadi dua jalur, yaitu jalur mahasiswa reguler dan jalur mahasiswa berprestasi.
2. Mahasiswa yang pernah mengikuti kompetisi tingkat nasional bisa langsung mengambil sertifikat tanpa harus mengantri.


```python
class AntrianSertifikat:
    def __init__(self, kapasitas=20):
        self.antrian = []
        self.antrian_reguler = []
        self.antrian_prestasi = []
        self.kapasitas = kapasitas

    def enqueue(self, nomor_antrian, prestasi=False):
        if prestasi:
            print(f"Mahasiswa {nomor_antrian} (prestasi) langsung mengambil sertifikat.")
        elif self.size() < self.kapasitas:
            self.antrian.append(nomor_antrian)
            print(f"Mahasiswa {nomor_antrian} masuk antrian.")
        else:
            if len(self.antrian_reguler) + len(self.antrian_prestasi) < self.kapasitas:
                if len(self.antrian_reguler) <= len(self.antrian_prestasi):
                    self.antrian_reguler.append(nomor_antrian)
                    print(f"Mahasiswa {nomor_antrian} masuk antrian reguler.")
                else:
                    self.antrian_prestasi.append(nomor_antrian)
                    print(f"Mahasiswa {nomor_antrian} masuk antrian prestasi.")
            else:
                print(f"Antrian penuh, mahasiswa {nomor_antrian} harus menunggu.")

    def dequeue(self):
        if not self.is_empty():
            mahasiswa_keluar = self.antrian.pop(0)
            print(f"Mahasiswa {mahasiswa_keluar} dilayani.")
        elif len(self.antrian_reguler) > 0:
            mahasiswa_keluar = self.antrian_reguler.pop(0)
            print(f"Mahasiswa {mahasiswa_keluar} (reguler) dilayani.")
        elif len(self.antrian_prestasi) > 0:
            mahasiswa_keluar = self.antrian_prestasi.pop(0)
            print(f"Mahasiswa {mahasiswa_keluar} (prestasi) dilayani.")
        else:
            print("Antrian kosong.")

    def front(self):
        if not self.is_empty():
            return self.antrian[0]
        elif len(self.antrian_reguler) > 0:
            return self.antrian_reguler[0]
        elif len(self.antrian_prestasi) > 0:
            return self.antrian_prestasi[0]
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.antrian[-1]
        elif len(self.antrian_prestasi) > 0:
            return self.antrian_prestasi[-1]
        elif len(self.antrian_reguler) > 0:
            return self.antrian_reguler[-1]
        else:
            return None

    def is_empty(self):
        return len(self.antrian) == 0 and len(self.antrian_reguler) == 0 and len(self.antrian_prestasi) == 0

    def size(self):
        return len(self.antrian) + len(self.antrian_reguler) + len(self.antrian_prestasi)

    def tampilkan_antrian(self):
        if not self.is_empty():
            print("Antrian mahasiswa:", self.antrian)
            print("Antrian reguler:", self.antrian_reguler)
            print("Antrian prestasi:", self.antrian_prestasi)
        else:
            print("Antrian kosong.")

# Fungsi untuk menghasilkan nomor antrian
def buat_nomor_antrian(prodi, npm):
    return prodi[0].upper() + npm[-2:]

# Contoh penggunaan dengan input pelanggan
antrian_sertifikat = AntrianSertifikat()

while True:
    print("\nMenu:")
    print("1. Tambah Mahasiswa ke Antrian")
    print("2. Layani Mahasiswa")
    print("3. Tampilkan Antrian")
    print("4. Keluar Program")

    pilihan = input("Masukkan pilihan Anda: ")

    if pilihan == "1":
        prodi = input("Masukkan program studi: ")
        npm = input("Masukkan NPM: ")
        prestasi = input("Apakah mahasiswa berprestasi (ya/tidak): ").lower() == "ya"
        nomor_antrian = buat_nomor_antrian(prodi, npm)
        antrian_sertifikat.enqueue(nomor_antrian, prestasi)
    elif pilihan == "2":
        antrian_sertifikat.dequeue()
    elif pilihan == "3":
        antrian_sertifikat.tampilkan_antrian()
    elif pilihan == "4":
        break
    else:
        print("Pilihan tidak valid.")

print("Program berakhir.")
```
