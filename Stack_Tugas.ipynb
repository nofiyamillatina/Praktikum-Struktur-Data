```python
class Gudang:
  def __init__(self, Kapasitas_gudang):
    self.barang = []
    self.Kapasitas_gudang = Kapasitas_gudang

  def push(self, nama_barang):
    if len(self.barang) >= self.Kapasitas_gudang:
      print("Gudang sudah penuh")
      return
    self.barang.append(nama_barang)
    print(f"{nama_barang} berhasil ditambahkan")

  def pop(self):
    self.barang.pop()
    if not self.barang:
      print("Barang tidak ditemukan")
      return
    else:
      print(f"{nama_barang} berhasil dihapus")

  def tampilkan(self):
    print("isi gudang")
    if not self.barang:
      print("Barang tidak ditemukan")
    else:
      for barang in reversed(self.barang):
        print(f"{barang}")

try:
  Kapasitas_gudang = int(input("Masukkan2 digit terakhir NPM anda: "))
  gudang = Gudang(Kapasitas_gudang)

  while True:
    print("\nMenu:")
    print("1. Tambah Barang")
    print("2. Hapus Barang")
    print("3. Tampilkan Barang")
    print("4. Keluar")

    pilihan = input("Pilih menu (1/2/3/4): ")

    if pilihan == "1":
      nama_barang = input("Masukkan nama barang: ")
      gudang.push(nama_barang)
    elif pilihan == "2":
      gudang.pop()
    elif pilihan == "3":
      gudang.tampilkan()
    elif pilihan == "4":
      print("Terima Kasih!")
      break
    else:
      print("Pilihan tidak valid. Silahkan coba lagi.")
except ValueError:
  print("Masukkan angka yang valid.")
```
